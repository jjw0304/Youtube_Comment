```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
chrome_driver_path = r'C:\Users\jjw03\Downloads\chromedriver.exe'  # chrome driver의 경로
driver = selenium.webdriver.Chrome(chrome_driver_path)
data = []

driver.get('https://www.youtube.com/watch?v=HbST1SnRTos') # 유튜브 영상의 링크

elem = driver.find_element_by_tag_name("body") 
pagedowns = 1
while pagedowns < 20: # 스크롤을 내릴 횟수
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(1)
        pagedowns += 1

html = driver.page_source
soup = BeautifulSoup(html)
table = soup.find('ytd-item-section-renderer', {'class' : 'style-scope ytd-comments'})
```
